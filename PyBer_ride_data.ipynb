```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Load data
pyber_ride_df = pd.read_csv("Resources/PyBer_ride_data.csv")
pyber_ride_df
```

```python
# Plot data frame
pyber_ride_df.plot(x="Month",y="Avg. Fare ($USD)")
plt.show()
```

```python
Graph = pyber_ride_df.plot(x="Month",y="Avg. Fare ($USD)")
Graph
```

```python
# Set x-axis and tick locations.
x_axis = np.arange(len(pyber_ride_df))
tick_locations = [value for value in x_axis]
# Plot the data.
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)")
plt.xticks(tick_locations, pyber_ride_df["Month"])
plt.show()
```

```python
# Making a bar chart --> multiple approaches!
pyber_ride_df.plot.bar(x="Month", y="Avg. Fare ($USD)")
plt.show()
```

```python
pyber_ride_df.plot.barh(x="Month", y="Avg. Fare ($USD)")

plt.gca().invert_yaxis()

plt.show()
```

```python
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)", kind='bar')
plt.show()
```

```python
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)", kind='barh')
plt.show()
```

```python
# Practice Exercise
SD = pyber_ride_df["Avg. Fare ($USD)"].std()
SD
```

```python
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)", kind='bar', yerr=SD, capsize=3, color = "skyblue")
plt.xticks(rotation = 0)

plt.yticks(np.arange(0,51,step=5))
plt.show()
```
