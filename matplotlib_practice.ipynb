```python
# Annotate charts by MATLAB - like interface
# Creat plot and label for legend
plt.plot(x_axis, y_axis, label = "Boston")
# X and Y axis labels
plt.xlabel("Month")
plt.ylabel("Fare in $")
# Set Y axis limits
plt.ylim(0,45)
# Title
plt.title("PyBer Fare by Month")
# Legend
plt.legend()
```

```python
# Change line and chart attributes
plt.plot(x_axis, y_axis, marker = "*", color = "blue", linewidth=2, label = "Boston")

plt.xlabel("Month")
plt.ylabel("Fare in $")
plt.ylim(0,45)
plt.title("PyBer Fare by Month")
plt.legend()

#Add a grid
plt.grid()
```

```python
# Chart annotation using object oriented interface
AwesomeGraph, ax = plt.subplots()
ax.plot(x_axis, y_axis, marker = "*", color = "green", linewidth=2, label = "Boston")
ax.set_ylim(0,45)
ax.set_xlabel('Month')
ax.set_ylabel("Fare in $")
ax.set_title("PyBer Fare by Month")
ax.legend()
ax.grid()
```

#### Bar Graphs

```python
# Data
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create bar chart using MATLAB-like interface
plt.bar(x_axis, y_axis)
```

```python
# Bar chart annotation
plt.bar(x_axis, y_axis, color = "green", label = "Boston")
plt.xlabel("Month")
plt.ylabel("Fare in USD($)")
plt.title("PyBer Fare by Month")
plt.legend()
```

```python
# Horizontal Bar Chart
plt.barh(x_axis, y_axis)
```

```python
# Invert y axis
plt.barh(x_axis, y_axis)
# Chain gca (get current axis) and invert methods
plt.gca().invert_yaxis()
```

```python
# Practice annotations on horizontal bar chart
plt.barh(x_axis, y_axis, color = "magenta", label = "Boston")

plt.gca().invert_yaxis()
plt.xlabel("Fare in USD($)")
plt.ylabel("Month")
plt.title("PyBer Fare by Month")
plt.legend()
```

#### Scatter Plots and Bubble Charts!

```python
# Scatter Plot using MATLAB like method

#Data:
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis, "o")
```

```python
plt.scatter(x_axis, y_axis)
```

```python
# Practice annotating with MATLAB method
plt.scatter(x_axis, y_axis, color = "red", label = "Chicago")
plt.xlabel("Month")
plt.ylabel("Fare in USD($)")
plt.title("PyBer Fares vs Month")
plt.legend()

plt.grid(5)
```

```python
# Bubble Charts

plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
# Create a new list to scale y-axis values by 3:
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
```

```python
# Plot with larger y-axis values
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# Scale within the size argument using list comprehension technique:
plt.scatter(x_axis, y_axis, s=[i*3 for i in y_axis])
```

```python
# Create scatter and bubble plots using object oriented interface

Graph, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# Practice changing attributes:
Graph2, ax = plt.subplots()
ax.scatter(y_axis, x_axis, color = "skyblue", alpha = 0.8, edgecolor = "black", linewidth = 2, s=[i*5 for i in y_axis], label = "Boston")
ax.invert_yaxis()

ax.set_xlabel("Fare in USD ($)")
ax.set_ylabel("Month")
ax.set_title("PyBer Fare vs Month")
ax.legend()
```

#### Pie Charts

```python
# Making Pie Charts using MATLAB method

plt.pie(y_axis, labels = x_axis)
plt.show()
# Notice that x_axis values were $ amounts, but are now converted to percentage of the pie for each month!
```

```python
# This makes the pie larger
plt.subplots(figsize=(8,8))

# This explodes/brings out a wedge
explode_values = (0,0,0,0,0,0, 0.2,0,0,0,0,0)

# autopct formats numbers to %
plt.pie(y_axis, explode = explode_values, labels = x_axis, autopct='%0.1f%%')
plt.show()
```

```python
# Assign each color:
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
# Making Pie Chart using OO Interface

Pie, ax = plt.subplots()
ax.pie(y_axis, labels = x_axis)
plt.show()
```

```python
# Format the OO pie chart:
Pie, ax = plt.subplots(figsize=(8, 8))
colorset = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
ax.pie(y_axis, labels = x_axis, explode = explode_values, colors = colorset, autopct='%0.1f%%', shadow = True, startangle=90)

ax.invert_xaxis()

plt.show()
```
